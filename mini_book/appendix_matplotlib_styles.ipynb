# Matplotlib styling

To install custom theme:

1. create the stylelib folder (if it does not already exist there) in your matplotlib directory. The matplotlib directory is typically in your user directory. To find it quickly, you can also just print it out in Python / Notebook with 

        import matplotlib as mpl
        mpl.get_configdir()

2. Put some custom styles (suffix *.mplstyle in your stylelib folder). Some templates can be found here https://github.com/DHI/matplotlib-styles .
3. Done! Now you can use the styles contained in the stylelib folder (e.g. DHI or DHI-wide from the templates) in your notebooks and scripts.
 
        import matplotlib.pyplot as plt
        plt.style.use('DHI')

The plots are then automatically in the selected format.


```python
# check where styles are located
import matplotlib as mpl
mpl.get_configdir()
```

```python
# import some data
import pandas as pd
df = pd.read_csv("data/oceandata.csv", parse_dates=True, index_col='datetime')

# plot without additional styling
df.plot()
```

```python
import matplotlib.pyplot as plt
plt.style.use('DHI-wide')
df.plot()
```
